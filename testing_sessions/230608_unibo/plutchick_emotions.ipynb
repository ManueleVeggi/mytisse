# Visualization of the emotions

Use of the library [PyPlutchik](https://github.com/alfonsosemeraro/pyplutchik). The file [```requirements.txt```](testing_sessions/230608_unibo/requirements.txt)contains the needed packages to run.

````python
from pyplutchik import plutchik
import pandas as pd

def emotion_data(dataDf, question, trustNaN=None):
    anger = dataDf['{0}_anger'.format(question)].astype(float).mean()
    disgust = dataDf['{0}_disgust'.format(question)].astype(float).mean()
    fear = dataDf['{0}_fear'.format(question)].astype(float).mean()
    interest = dataDf['{0}_interest'.format(question)].astype(float).mean()
    joy = dataDf['{0}_joy'.format(question)].astype(float).mean()
    sadness = dataDf['{0}_sadness'.format(question)].astype(float).mean()
    surprise = dataDf['{0}_surprise'.format(question)].astype(float).mean()
    trust = dataDf['{0}_trust'.format(question)].astype(float).mean()

    feelings = {
        "anger": round(anger / 5, 2), 
        "disgust": round(disgust / 5, 2), 
        "fear": round(fear / 5, 2), 
        "anticipation": round(interest / 5, 2), 
        "joy": round(joy / 5, 2),
        "sadness": round(sadness / 5, 2),
        "surprise": round(surprise / 5, 2),
        "trust": round(trust / 5, 2)}
    
    if trustNaN:
        feelings['trust'] = 0
    
    return feelings    

mytisse_data = pd.read_csv("data/data_cleaned.csv", sep=";")
groupA = mytisse_data[mytisse_data['group_id'] == 'A']
groupB = mytisse_data[mytisse_data['group_id'] == 'B']
groupC = mytisse_data[mytisse_data['group_id'] == 'C']

````

## Overall values

````python
# Bathers by A River

plutchik(emotion_data(mytisse_data, "q4"))
````

````python
# Edit 1

plutchik(emotion_data(mytisse_data, "q10"))
````

````python
# Edit 1

plutchik(emotion_data(mytisse_data, "q12", True))
````

## Group A

````python
plutchik(emotion_data(groupA, "q4"), title="Original")
plutchik(emotion_data(groupA, "q10"), title="Edit 1")
plutchik(emotion_data(groupA, "q12", True), title="Edit 2")
````

## Group B

````python
plutchik(emotion_data(groupB, "q4"), title="Original")
plutchik(emotion_data(groupB, "q10"), title="Edit 1")
plutchik(emotion_data(groupB, "q12", True), title="Edit 2")
````

## Group C

````python
plutchik(emotion_data(groupC, "q4"), title="Original")
plutchik(emotion_data(groupC, "q10"), title="Edit 1")
plutchik(emotion_data(groupC, "q12", True), title="Edit 2")
````
